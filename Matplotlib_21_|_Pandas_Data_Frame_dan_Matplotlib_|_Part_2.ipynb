## **Matplotlib 21 | Pandas Data Frame dan Matplotlib | Part 2 | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

### **Line Plot**

Sample Dataset

```python
data = {
    'Tahun': [
        '1958', '1963', '1968', '1973', '1978', '1983', '1988', '1993', '1998', '2003', '2008', '2013', '2018'],
    'Populasi': [51652500, 53624900, 55213500, 56223000, 56178000, 56315000, 56916000, 57713000, 58474000, 59636000, 61823000, 64105000, 66436000]
}

df = pd.DataFrame(data)
df
```

```python
df.plot(x='Tahun', y='Populasi', kind='line')

plt.title('Data Populasi Penduduk Indonesia')
plt.ylabel('Populasi (dalam Juta Orang)')
plt.xlabel('Tahun')
plt.show()
```

### **Bar Plot**

Sample Dataset

```python
data = {
    'Negara': ['United States', 'Singapore', 'Germany', 'United Kingdom', 'Japan'],
    'GDP': [62606, 100345, 52559, 45705, 44227]
}

df = pd.DataFrame(data, columns=['Negara', 'GDP'])
df
```

```python
df.plot(x='Negara', y='GDP', kind='bar')

plt.title('GDP Per Kapita')
plt.ylabel('GDP ($)')
plt.xlabel('Negara')
plt.show()
```

```python
df.plot(x='Negara', y='GDP', kind='barh')

plt.title('GDP Per Kapita')
plt.ylabel('Negara')
plt.xlabel('GDP ($)')
plt.show()
```

```python
df.plot(x='Negara', y='GDP', kind='barh', color='g', legend=False)

plt.title('GDP Per Kapita')
plt.ylabel('Negara')
plt.xlabel('GDP ($)')
plt.show()
```

### **Scatter Plot**

Sample Dataset

```python
data = {
    'Negara': ['United States', 'Singapore', 'Germany', 'United Kingdom', 'Japan'],
    'GDP': [62606, 100345, 52559, 45705, 44227],
    'Life_Expetancy': [79.24, 82.84, 80.84, 81.40, 83.64]
}

df = pd.DataFrame(data, columns=['Negara', 'GDP', 'Life_Expetancy'])
df
```

```python
df.plot(kind='scatter', x='GDP', y='Life_Expetancy', color='r',
        title='GDP and Life Expetancy', ylabel='Life Expetancy', xlabel='GDP')

plt.show()
```
