## **Matplotlib 21 | Pandas Data Frame dan Matplotlib | Part 2 | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

### **Line Plot**

Sample Dataset

```python
data = {
    'Tahun': [
        '1958', '1963', '1968', '1973', '1978', '1983', '1988', '1993', '1998', '2003', '2008', '2013', '2018'],
    'Populasi': [51652500, 53624900, 55213500, 56223000, 56178000, 56315000, 56916000, 57713000, 58474000, 59636000, 61823000, 64105000, 66436000]
}

df = pd.DataFrame(data)
df
```

```python
df.plot(x='Tahun', y='Populasi', kind='line')

plt.title('Data Populasi Penduduk Indonesia')
plt.ylabel('Populasi (dalam Juta Orang)')
plt.xlabel('Tahun')
plt.show()
```
