## **Matplotlib 16 | Alur Hidup Plot | Plot Lifecycle | Belajar Matplotlib Dasar | Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
print(pd.__version__)
```

### **Sample Dataset**

```python
data = {'Item A': 109438,
'Item B': 103569,
'Item C': 112214,
'Item D': 112591,
'Item E': 100934,
'Item F': 103660,
'Item G': 137351,
'Item H': 123381,
'Item I': 135841,
'Item J': 104437}

data
```

```python
items = tuple(data.keys())
items
```

```python
data.keys()
```

```python
count = tuple(data.values())
count
```

```python
data.values()
```

### **Simple Plot**

```python
fig, ax = plt.subplots()

ax.barh(items, count)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(items, count)

plt.show()
```

```python
fig, ax = plt.subplots()

# The 'count' tuple already contains the numeric values needed for the pie chart.
# Simply pass 'count' directly to the 'ax.pie' function.
# The 'autopct' parameter is added to display percentages on the pie slices.
ax.pie(count, labels=items, autopct='%1.1f%%')

plt.show()
```

## **Pengaturan Style**

```python
plt.style.available
```

```python
plt.style.use('bmh')

fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

### **Pengaturan Tick Label**

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

# set properties
plt.setp(labels,
         rotation=45,
         horizontalalignment='right')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels()

# set properties
plt.setp(labels,
         rotation=45,
         horizontalalignment='left')

plt.show()
```

### **Pengaturan Format pada Ticker**

```python
from matplotlib.ticker import FuncFormatter

def ribuan(x, pos):
  return f'{int(x/1000)}K'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

### **Pengaturan Label pada Sumbu (axis) dan Judul (title)**

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```
