# **Matplotlib 14: Belajar Contour Plot | Contour Chart | Belajar Matplotlib Dasar | Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### **Sample Dataset**

```python
n = np.linspace(-3.0, 3.0, 100)

x, y = np.meshgrid(n, n)
z = np.sqrt(x**2 + y**2)

print(x.shape)
print(y.shape)
print(z.shape)
```

```python
x
```

```python
y
```

```python
z
```

### **Contour Plot**

```python
plt.contour(x, y, z)
plt.colorbar()

plt.title('Countour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```

### **Filled Countour Plot**

```python
plt.coutour(x, y, z)
plt.colorbar()

plt.title('Filled Countour Plot')
plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')

plt.show()
```
