### **Matplotlib 11: Belajar Violin Plot | Violin Chart | Belajar Matplotlib Dasar | Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### **Sample Dataset**

```python
np.random.seed(2)
data = np.random.normal(loc=100, scale=10, size=200)
data
```

### **Simple Violin Plot**

```python
plt.violinplot(data)

plt.title('Simple Violin Plot')
plt.xlabel('Data')
plt.ylabel('Value')

plt.show()
```

### **Pengaturan pada Violin Plot**

```python
plt.violinplot(data,
               showextrema=False,
               showmeans=False,
               showmedians=False,
               quantiles=None)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu X')
plt.xlabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=True,
               showmeans=False,
               showmedians=False,
               quantiles=None)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu X')
plt.xlabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=False,
               showmeans=True,
               showmedians=False,
               quantiles=None)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu X')
plt.xlabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=False,
               showmeans=False,
               showmedians=True,
               quantiles=None)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu X')
plt.xlabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=True,
               showmeans=True,
               showmedians=True,
               quantiles=None)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu X')
plt.xlabel('Sumbu Y')

plt.show()
```

```python
plt.violinplot(data,
               showextrema=False,
               showmeans=False,
               showmedians=True,
               quantiles=None)

plt.title('Simple Violin Plot')
plt.ylabel('Sumbu X')
plt.xlabel('Sumbu Y')

plt.show()
```
