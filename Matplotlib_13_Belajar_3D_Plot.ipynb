# **Matplotlib 13: Belajar 3D Plot | 3D Chart | Belajar Matplotlib Dasar | Belajar Visualisasi Data**



```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### **3D Axes**

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

plt.show()
```

### **3D Line PLot**

```python
z = np.linspace(0, 1, 100)
x = z * np.sin(20 * z)
y = z * np.cos(20 * z)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='r')

ax.set(title='Contoh 3D Line Plot',
       xlabel='x',
       ylabel='y',
       zlabel='z')

plt.show()
```
