# **Matplotlib 06: Belajar Pie Plot | Pie Chart | Belajar Matplotlib Dasar | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

### **Simple Pie Plot**

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 105]
```

```python
plt.pie(data,
        labels=kategori,
        autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Pie Plot')
plt.show()
```

```python
plt.pie(data,
        # labels=kategori,
        # autopct='%1.1f%%',
        startangle=90)

plt.title('Contoh Pie Plot')
plt.show()
```
