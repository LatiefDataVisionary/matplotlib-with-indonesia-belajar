## **Matplotlib 20 | Pandas Data Frame dan Matplotlib | Part 1 | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

### **Sample Dataset**

```python
df = pd.DataFrame({
'nama': ['cecep', 'karti', 'bejo', 'tejo', 'asep', 'wati', 'amat'],
'usia': [23, 78, 22, 19, 45, 33, 21],
'gender': ['L', 'P', 'L', 'L', 'L', 'P', 'L'],
'propinsi':['jabar', 'jatim', 'jabar', 'jatim', 'jabar', 'jateng', 'jateng'],
'n_anak': [2, 0, 0, 3, 2, 1, 4],
'n_pets': [5, 1, 0, 5, 2, 2, 3]
})

df
```

### **Scatter Plot**

```python
df.plot(kind='scatter', x='n_anak', y='n_pets', color='red')

plt.title('Scatter Plot')
plt.xlabel('Jumlah Anak')
plt.ylabel('Jumlah Kucing')

plt.show()
```

### **Histogram**

```python
df[['usia']].plot(kind='hist', bins=[0, 20, 40, 60, 80, 100])

plt.title('Histogram')
plt.xlabel('Usia')
plt.ylabel('Jumlah')
plt.show()
```
