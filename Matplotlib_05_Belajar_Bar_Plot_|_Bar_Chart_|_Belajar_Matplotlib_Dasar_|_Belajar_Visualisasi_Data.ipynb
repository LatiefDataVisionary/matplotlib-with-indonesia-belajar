# **Matplotlib 05: Belajar Bar Plot | Bar Chart | Belajar Matplotlib Dasar | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### **Simple Bar Plot**

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### **Pengaturan Grid dan Color**

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori,
        data,
        color = 'red',
        alpha=0.5)

plt.grid(linestyle='--',
         linewidth=2,
         axis='y', # pilihan: x, y, both
         alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

```python
plt.bar(kategori,
        data,
        color = 'red',
        alpha=0.5)

plt.grid(linestyle='--',
         linewidth=2,
         axis='both', # pilihan: x, y, both
         alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### **Horisontal Bar Plot**

```python
data = [25, 45, 55, 125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.barh(kategori, data)
# plt.barh(kategori[::-1], data[::-1])

plt.grid(linestyle='--',
         linewidth=1,
         axis='x',
         alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horizontal Bar Plot')

plt.show()
```

```python
plt.barh(kategori[::-1], data[::-1])

plt.grid(linestyle='--',
         linewidth=1,
         axis='x',
         alpha=0.75)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horizontal Bar Plot')

plt.show()
```

### **Stacked Bar Plot**

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1', color='red')
# plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data2, label='Data 2', color='b', bottom=data1, alpha=0.75)

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
plt.bar(kategori, data1, label='Data 1', color='red', bottom=data2)
# plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data2, label='Data 2', color='b', alpha=0.75)

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
plt.bar(kategori, data1, label='Data 1')
plt.bar(kategori, data2, label='Data 2')

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```

```python
plt.bar(kategori, data2, label='Data 2')
plt.bar(kategori, data1, label='Data 1')

plt.legend()

plt.grid(linestyle='--',
         linewidth=1,
         axis='y',
         alpha=0.75)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar Plot')

plt.show()
```
