# **Matplotlib 04: Belajar Histogram | Belajar Matplotlib Dasar | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### **Sample Data**

```python
mu, sigma = 100, 15  # Nilai mean dan nilai standard deviation

x = mu + sigma * np.random.randn(10000)  # Normal Distribution
x
```

```python
x.shape
```

### **Histogram dengan pyplot Style**

```python
plt.hist(x,
         bins=50,  # 10000/50
         facecolor='g',
         alpha=0.75) # Transparansi 75%

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100,\ \sigma=15$')
plt.grid()

plt.show()
```

### **Histogram dengan OO Style**

```python
fig, ax = plt.subplots()

ax.hist(x,
         bins=50,  # 10000/50
         facecolor='r',
         alpha=0.25) # Transparansi 75%

ax.set_xlabel('Sumbu X')
ax.set_ylabel('Sumbu Y')
ax.set_title('Contoh Histogram')

ax.text(45, 500, '$\mu=100,\ \sigma=15$')
ax.grid()

plt.show()
```
