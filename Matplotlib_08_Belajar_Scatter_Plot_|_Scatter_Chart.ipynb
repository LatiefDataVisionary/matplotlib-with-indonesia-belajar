# **Matplotlib 08: Belajar Scatter Plot | Scatter Chart | Belajar Matplotlib Dasar | Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

### **Simple Scatter Plot**

```python
x = [2,4,6,8,10,11,11.5,11.7]
y = [1,1.5,2,2.5,3,3.5,4,4.5]
```

```python
plt.scatter(x, y, label='Data 1', color='r')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')

plt.legend()

plt.show()
```

```python
x2, y2 = y, x

plt.scatter(x, y, label='Data 1', color='r')
plt.scatter(x2, y2, label='Data 2', color='g')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Scatter Plot')

plt.legend()

plt.show()
```

### **Multiple Scatter Plot**

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x1, y1, label='Data 1', color='r')
plt.scatter(x2, y2, label='Data 2', color='b')

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Multiple Scatter Plot')

plt.legend()
plt.show()
```

### **Pengaturan Marker**

```python
x1 = [2,4,6,8,10,11,11.5,11.7]
y1 = [1,1.5,2,2.5,3,3.5,4,4.5]

x2 = [8,8.5,9,9.5,10,10.5,11]
y2 = [3,3.5,3.7,4,4.5,5,5.2]
```

```python
plt.scatter(x1, y1,
            color='cyan',
            linewidths=1,
            marker='s',  # penanda
            edgecolor='red',  # warna garis tepi
            s=100) # size

plt.scatter(x2, y2,
            color='yellow',
            linewidths=1,
            marker='^',
            edgecolor='blue',
            s=200)

plt.show()
```
