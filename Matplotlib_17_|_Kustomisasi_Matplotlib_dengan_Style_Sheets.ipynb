## **Matplotlib 17 | Kustomisasi Matplotlib dengan Style Sheets | Matplotlib Dasar | Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

print(matplotlib.__version__)
print(np.__version__)
print(pd.__version__)
print(sns.__version__)
```

### **Predefined Style**

Matplotlib menyediakan beberapa predefined styles. Detail dari tiap predefined style bisa ki

```python
plt.style.available
```

```python
plt.style.use('ggplot')
```

```python
x = np.linspace(1, 10, 5)
```

```python
x
```

```python
np.log(x)
```

```python
plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')
plt.legend()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')

plt.show()
```

### **Membuat Style Sendiri**

`axes.titlesize : 24`

`axes.labelsize : 20  `

`lines.linewidth : 3 `

`lines.markersize : 10 `

`xtick.labelsize : 16 `

`ytick.labelsize : 16 `





```python
style_ku = {
    'axes.titlesize': 24,
    'axes.labelsize': 20,
    'lines.linewidth': 3,
    'lines.markersize': 10,
    'xtick.labelsize': 16,
    'ytick.labelsize': 16
}

plt.rcParams.update(style_ku) # Update rcParams with the custom style
```

```python
x = np.linspace(1, 10, 5)
```

```python
plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')
plt.legend()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Plot')

plt.show()
```

### **Menerapkan Multiple Style**

```python
sns.set_theme(style="darkgrid")  # Use 'darkgrid' for a dark background
```

```python
x = np.linspace(1, 10, 5)

plt.plot(x, x, 'bs-', label='lin')
plt.plot(x, np.log(x), 'ro-', label='log')
plt.legend()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Plot')

plt.show()
```

### **Menerapkan Temporary Styling**

```python
plt.style.use('ggplot')

plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')
plt.show()
```

```python
with plt.style.context('dark_background'):
  plt.plot(np.sin(np.linspace(0, 2 * np.pi)), 'r-o')

plt.show()
```
