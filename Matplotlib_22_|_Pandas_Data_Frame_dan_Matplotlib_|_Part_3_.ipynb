## **Matplotlib 22 | Pandas Data Frame dan Matplotlib | Part 3 | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

### **Sample Dataset**

```python
df = pd.read_csv('https://raw.githubusercontent.com/boedybios/belajar_matplotlib/master/dataset/2020.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.rename(columns = {
    'Country name': 'Country',
    'Perceptions of corruption': 'Corruption',
    'Freedom to make life choices': 'Freedom'
},
          inplace=True)

df.columns
```

### **Bar Plot**

```python
df[:5].plot(x='Country',
            y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
            kind='bar',
            title='Contoh Bar Plot',
            ylabel='Nilai')

plt.show()
```

### **Line Plot**

```python
df[:5].plot(x='Country',
            y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
            kind='line',
            title='Contoh Line Plot',
            ylabel='Nilai')

plt.show()
```

### **Box Plot**

```python
df.plot(x='Country',
            y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
            kind='box',
            title='Contoh Box Plot',
            xlabel='Parameter',
            ylabel='Nilai')

plt.show()
```

```python
df[:50].plot(x='Country',
            y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
            kind='box',
            title='Contoh Box Plot',
            xlabel='Parameter',
            ylabel='Nilai')

plt.show()
```

### **Scatter Plot**

```python
df.plot(x='Corruption',
        y='Freedom',
        kind='scatter',
        xlim=(0, 1),
        ylim=(0, 1),
        title='Contoh Scatter Plot',
        xlabel='Corruption',
        ylabel='Freedom')

plt.show()
```

```python
df[:50].plot(x='Corruption',
        y='Freedom',
        kind='scatter',
        xlim=(0, 1),
        ylim=(0, 1),
        title='Contoh Scatter Plot',
        xlabel='Corruption',
        ylabel='Freedom')

plt.show()
```

### **Histogram**

```python
df.plot(x='Country',
        y='Corruption',
        kind='hist',
        bins=10,
        title='Contoh Histogram',
        xlabel='Nilai')

plt.show()
```

```python
df.plot(x='Country',
        y='Corruption',
        kind='hist',
        bins=20,
        title='Contoh Histogram',
        xlabel='Nilai')

plt.show()
```

```python
df.plot(x='Country',
        y='Corruption',
        kind='hist',
        bins=5,
        title='Contoh Histogram',
        xlabel='Nilai')

plt.show()
```

```python
df.plot(x='Country',
        y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
        kind='hist',
        subplots=True,
        layout=(2, 2))

plt.suptitle('Contoh Subplots')
plt.tight_layout()
plt.show()
```
