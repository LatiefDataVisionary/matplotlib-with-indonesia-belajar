## **Matplotlib 22 | Pandas Data Frame dan Matplotlib | Part 3 | Belajar Visualisasi Data**


```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd

print(matplotlib.__version__)
print(pd.__version__)
```

### **Sample Dataset**

```python
df = pd.read_csv('https://raw.githubusercontent.com/boedybios/belajar_matplotlib/master/dataset/2020.csv')
df.head(10)
```

```python
df.rename(columns = {
    'Country name': 'Country',
    'Perceptions of corruption': 'Corruption',
    'Freedom to make life choices': 'Freedom'
},
          inplace=True)

df.columns
```

### **Bar Plot**

```python
df[:5].plot(x='Country',
            y=['Corruption', 'Freedom', 'Generosity', 'Social support'],
            kind='bar',
            title='Contoh Bar Plot',
            ylabel='Nilai')

plt.show()
```
